#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

```python
valdemorillo = df[df['level5'] == 'Valdemorillo']
galapagar = df[df['level5'] == 'Galapagar']
av_pr_valdemorillo=round(valdemorillo['price'].mean(), 2)
av_pr_galapagar=round(galapagar['price'].mean(), 2)

print(f'''Average House Prices in Valdemorillo {av_pr_valdemorillo}
Average House Prices in Galapagar {av_pr_galapagar}''')
if av_pr_galapagar==av_pr_valdemorillo:
    print('the avg prices of housing are the same')
elif av_pr_galapagar > av_pr_valdemorillo:
    print('Galapagar is a more expensive zone to live')
else:
    print('Valdemorillo is a more expensive zone to live')
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_southbelt['precios_normalizados'] = scaler.fit_transform(df_southbelt[['price']])
plt.figure(figsize=(10, 6))

data = []
print(df_southbelt['level5'].unique())
for ciudad in df_southbelt['level5'].unique():
    subset = df_southbelt[df_southbelt['level5'] == ciudad]
    data.append(subset['precios_normalizados'])

plt.hist(data, bins=30, stacked=False, label=df_southbelt['level5'].unique())

plt.legend(loc='upper right')
plt.title('Precios Normalizados para Cada Población')
plt.xlabel('Precios Normalizados')
plt.ylabel('Frecuencia')
plt.show()
```

Fuenlabrada: The distribution is centered around zero, indicating that most prices are close to the mean after normalization.

Leganés: The distribution is similar to Fuenlabrada, but with a narrower spread. The majority of prices are still close to the mean, and there are fewer extreme values.

Getafe: The distribution is more symmetric, with a peak around zero. Prices are well-clustered, indicating less variability after normalization.

Alcorcón: The distribution is positively skewed, with a long tail on the right side. This suggests that Alcorcón has a few significantly higher-priced items compared to the other populations.

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, Marker, Icon
import pandas as pd
df_southbelt['latitude'] = df_southbelt['latitude'].astype(str).str.replace(',', '.')
df_southbelt['longitude'] = df_southbelt['longitude'].astype(str).str.replace(',', '.')
df_southbelt['latitude'] = df_southbelt['latitude'].astype(float)
df_southbelt['longitude'] = df_southbelt['longitude'].astype(float)
cities = df_southbelt['level5'].unique()
colors = ['red', 'blue', 'green', 'yellow', 'orange']
city_color_map = dict(zip(cities, colors))
madrid_map = Map(center=(40.3253, -3.7878), zoom=11)
for city in cities:
    subset = df_southbelt[df_southbelt['level5'] == city]
    for _, row in subset.iterrows():
        marker = Marker(
            location=(float(row['latitude']), float(row['longitude'])),
            icon=Icon(icon_url=f'http://maps.google.com/mapfiles/ms/icons/{city_color_map[city]}-dot.png')
        )
        madrid_map.add_layer(marker)
madrid_map
```
